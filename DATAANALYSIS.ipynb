```python
import matplotlib.pyplot as plt

# Histogram for 'ATTA'
plt.figure()
plt.hist(df['ATTA '], bins=10)
plt.xlabel('ATTA')
plt.ylabel('Frequency')
plt.title('Histogram of ATTA')
plt.show()

# Scatter plot for 'ATTA' vs 'GHEE'
plt.figure()
plt.scatter(df['ATTA '], df['GHEE '])
plt.xlabel('ATTA')
plt.ylabel('GHEE')
plt.title('Scatter Plot of ATTA vs GHEE')
plt.show()

# Box plot for all columns
plt.figure()
df.boxplot()
plt.title('Box Plot of All Columns')
plt.show()
```

```python
# prompt: generate different graphs

# Pair plot for all numerical columns
sns.pairplot(df)
plt.show()

# Violin plot for 'ATTA'
plt.figure()
sns.violinplot(x=df['ATTA '])
plt.title('Violin Plot of ATTA')
plt.show()

# Heatmap for correlation matrix
#corr_matrix = df.corr()
#plt.figure()
#sns.heatmap(corr_matrix, annot=True)
#plt.title('Heatmap of Correlation Matrix')
#plt.show()

# Line plot for 'ATTA' over time (assuming 'Date' column exists)
if 'Date' in df.columns:
  df_sorted = df.sort_values('Date')
  plt.figure()
  plt.plot(df_sorted['Date'], df_sorted['ATTA '])
  plt.xlabel('Date')
  plt.ylabel('ATTA')
  plt.title('Line Plot of ATTA over Time')
  plt.xticks(rotation=45)
  plt.show()
```
